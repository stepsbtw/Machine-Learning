## Curva de Aprendizado

Plot que mostram como a performance de um modelo de machine learning (accuracy, error rate) muda em função do tamanho do dataset de treino.

- Úteis para analisar **underfitting e overfitting**.

### Overfitting (alto viés, baixa variância)

O modelo "memorizou" o conjunto de treino. Mesmo aumentando o dataset não sou capaz de diminuir o erro de validação.

image.png

### UnderFitting (alta variância, baixo viés)

Nesse caso, o modelo pode sim se beneficiar do aumento do dataset de treino, mas talvez o modelo seja muito simples.

image.png

### Ideal

Talvez já esteja em platô, mas não está enviesado nos dados de treino! Consegue um bom resultado nos dados de validação.
