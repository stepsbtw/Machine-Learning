# Tipos de Aprendizado
Como os dados so apresentados para o modelo?
- Supervisionado (Inductive Learning)
  - Dados rotulados (entrada e sada)
  - Regresso Linear, Logstica, Classificao de Imagens
- No Supervisionado
  - Dados no rotulados (Padres e regras OCULTAS)
  - kMeans, PCA (reduo de dimensionalidade), Autoencoders
- Reinforcement Learning
  - Um agente interagindo com ambiente (Recompensas e punies)
  - Agentes em jogos (Pacman)

- Semi-Supervisionado
- Self-Supervisionado

# Supervised Learning

- Encontrar uma aproximao de uma funo f(x) apartir de exemplos de f.
- f(x)  nossa **funo objetivo**

- Input: Conjunto de exemplos T {(x,f(x))}
- Output: Funo h(x) a qual  uma aproximao de f(x) (Funo **Hiptese**)

### Hiptese
Uma funo determinada pelo **algoritmo de aprendizado** que mapeia:
x -> aproximao de f(x)

## Classificao x Regresso
O aprendiz recebe as respostas no aprendizado supervisionado.

- Classificao : Predizer um valor discreto
- Regresso : Predizer um valor contnuo

# Componentes de Algoritmos de Aprendizado
Todo algoritmo de aprendizado de mquina contm 3 componentes:

1. **Representao**: Qual "lnguagem"  utilizada para representar os programas (as hipteses) que o algoritmo pode aprender.
  - rvores de Deciso
  - Redes Neurais Artificiais
  - Instncias (Apenas armazenam os dados)
  - Ensembles (Conjuntos de Modelos)
2. **Avaliao**: Como a qualidade de um programa  objetivamente avaliado.
  - Acurcia (Proporo de previses corretas)
  - Preciso = TP / (TP + FP), Recall = TP / (TP + FN), F1 = mdia harmnica
  - Erros Quadrticos
  - Verossimilhana
3. **Otimizao**: Como os programas candidatos so gerados.
  - Combinatria : ex - Greed Search
  - Convexa : ex - Descida de Gradiente
  - Com Restries : ex - Programao Linear
