# Overview

Field of study that gives computers ability **to learn without explicitly programming**.

image.png


*The Master Algorithm* - Pedro Domingos

Diferentes escolas de pensamento de modelos:
- Symbolists
  - Baseados em regra (Lógica e Programação Indutiva)
  - Representação explícita de conhecimento e raciocínio lógico
  - Ex: Árvores de Decisão
- Evolutionists
  - Algoritmos Genéticos
  - Otimização por seleção natural
  - Solucões adaptativas ao longo de gerações
- Analogists
  - Baseados em Instâncias (Semelhança com exemplos anteriores)
  - kNNs, SVMs, Regressão Linear, Logística
- Bayesians
  - Probabilidade e atualização Bayesiana
  - Redes Bayesianas e Inferência Estatística
  - Ex: Naive Bayes
  - Bayesian Optimization (Hiperparâmetros)
- Connectionists
  - Redes Neurais Artificiais
  - Deep Learning
  - Modelagem de redes neurais inspiradas no cérebro humano

# Tipos de Aprendizado
Como os dados são apresentados para o modelo?
- Supervisionado (Inductive Learning)
  - Dados rotulados (entrada e saída)
  - Regressão Linear, Logística, Classificação de Imagens
- Não Supervisionado
  - Dados não rotulados (Padrões e regras OCULTAS)
  - kMeans, PCA (redução de dimensionalidade), Autoencoders
- Reinforcement Learning
  - Um agente interagindo com ambiente (Recompensas e punições)
  - Agentes em jogos (Pacman)

- Semi-Supervisionado
- Self-Supervisionado

# Supervised Learning

- Encontrar uma aproximação de uma função f(x) apartir de exemplos de f.
- f(x) é nossa **função objetivo**

- Input: Conjunto de exemplos T {(x,f(x))}
- Output: Função h(x) a qual é uma aproximação de f(x) (Função **Hipótese**)

### Hipótese
Uma função determinada pelo **algoritmo de aprendizado** que mapeia:
x -> aproximação de f(x)

## Classificação x Regressão
O aprendiz recebe as respostas no aprendizado supervisionado.

- Classificação : Predizer um valor discreto
- Regressão : Predizer um valor contínuo

# Componentes de Algoritmos de Aprendizado
Todo algoritmo de aprendizado de máquina contém 3 componentes:

1. **Representação**: Qual "línguagem" é utilizada para representar os programas (as hipóteses) que o algoritmo pode aprender.
  - Árvores de Decisão
  - Redes Neurais Artificiais
  - Instâncias (Apenas armazenam os dados)
  - Ensembles (Conjuntos de Modelos)
2. **Avaliação**: Como a qualidade de um programa é objetivamente avaliado.
  - Acurácia (Proporção de previsões corretas)
  - Precisão = TP / (TP + FP), Recall = TP / (TP + FN), F1 = média harmônica
  - Erros Quadráticos
  - Verossimilhança
3. **Otimização**: Como os programas candidatos são gerados.
  - Combinatória : ex - Greed Search
  - Convexa : ex - Descida de Gradiente
  - Com Restrições : ex - Programação Linear
